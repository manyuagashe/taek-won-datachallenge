```python

def flatten_list(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):  # If it's a nested list, extend the flat list with its items
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)  # If it's not a list, just append the item
    return flat_list

# Flatten the nested synonyms list
goodsynonyms_hashed = flatten_list(goodsynonyms)

# Now you can create a se
my_set = set(goodsynonyms_hashed)
len(my_set)
```

```python
fullclubdf = pd.concat([dfclub,dfbar,dfdisco,dfpubfilter,dfparty])
fullclubdffiltered=fullclubdf[fullclubdf['numReviews']>3.0]
fullclubeurope = fullclubdffiltered[fullclubdffiltered['location']!="Dubai"]
fullclubeurope
```

```python
fullclubeurope = fullclubeurope.rename(columns={"name": "poi_name"})
```

```python

review_dfs = []  

for index, row in fullclubeurope.iterrows():
    review_api_url = row['reviews'] 
    
    response = requests.get(review_api_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        review_data = response.json() 
        
        
        if isinstance(review_data, list):
            review_df = pd.DataFrame(review_data)
            review_df['poi_name'] = row['poi_name'] 
            
            # Append the individual review DataFrame to the list
            review_dfs.append(review_df)
        else:
            print(f"No reviews available for POI: {row['poi_name']}")
    else:
        print(f"Failed to retrieve reviews for POI: {row['poi_name']} (Status: {response.status_code})")
if review_dfs:
    all_reviews_df = pd.concat(review_dfs, ignore_index=True)
all_reviews_df.head()
```

```python
all_reviews_dfen = all_reviews_df[all_reviews_df['language']=='en']
all_reviews_dfen
```

```python
reviewlist=all_reviews_dfen['text'].tolist()
reviewlist
```

```python
def check_for_synonyms(review, synonyms):
        review_lower = review.lower()
        return any(synonym.lower() in review_lower for synonym in synonyms)
goodresult = [check_for_synonyms(review, goodsynonyms_hashed) for review in reviewlist]
sum(goodresult)
```

```python
badresult = [check_for_synonyms(review, badsynonyms_hashed) for review in reviewlist]
sum(badresult)
```

```python
all_reviews_dfen.insert(9,'safe',goodresult)
```

```python
all_reviews_dfen.insert(10,'unsafe',badresult)
```

```python
mergedf=pd.merge(all_reviews_dfen,fullclubeurope,on='poi_name',how='left')
```

```python
mergedf
```

```python
series_1=mergedf.groupby(['location'])["safe"].mean()-mergedf.groupby(['location'])["unsafe"].mean()
series_1
```

```python
series_2=mergedf.groupby(['location'])['polarity_x'].mean()
```

```python
correlation = series_1.corr(series_2)
correlation
```

```python
safetyperpoi=mergedf.groupby(["poi_name"])["safe"].mean()-mergedf.groupby(["poi_name"])["unsafe"].mean()
```

```python
polarityperpoi=mergedf.groupby(["poi_name"])["polarity_x"].mean()
print(polarityperpoi)
```

```python


norm = plt.Normalize(polarityperpoi.min(), polarityperpoi.max())
cmap = plt.get_cmap("coolwarm")  

plt.figure(figsize=(10, 6))
scatter = plt.scatter(safetyperpoi, polarityperpoi, c=polarityperpoi, cmap=cmap, s=100, edgecolor='black')
plt.colorbar(scatter, label='Polarity')

plt.title('Safety vs Polarity with Gradient per POI', fontsize=16)
plt.xlabel('Safety Score')
plt.ylabel('Polarity Score')

plt.show()
```
