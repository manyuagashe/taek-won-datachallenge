```python

def flatten_list(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):  # If it's a nested list, extend the flat list with its items
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)  # If it's not a list, just append the item
    return flat_list

# Flatten the nested synonyms list
goodsynonyms_hashed = flatten_list(goodsynonyms)

# Now you can create a se
my_set = set(goodsynonyms_hashed)
len(my_set)
```

```python
fullclubdf = pd.concat([dfclub,dfbar,dfdisco,dfpubfilter,dfparty])
fullclubdffiltered=fullclubdf[fullclubdf['numReviews']>3.0]
fullclubeurope = fullclubdffiltered[fullclubdffiltered['location']!="Dubai"]
fullclubeurope
```

```python
fullclubeurope = fullclubeurope.rename(columns={"name": "poi_name"})
```

```python
# Assume `df` is your main DataFrame with a 'reviews' column containing API links
review_dfs = []  # List to store individual DataFrames of reviews

# Iterate through the rows of the DataFrame
for index, row in fullclubeurope.iterrows():
    review_api_url = row['reviews']  # Get the review API link for the current POI
    
    # Make a GET request to fetch the reviews from the API
    response = requests.get(review_api_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        review_data = response.json()  # Parse the JSON response
        
        # If the review data is a list, convert it into a DataFrame
        if isinstance(review_data, list):
            review_df = pd.DataFrame(review_data)
            review_df['poi_name'] = row['name']  # Add the name of the POI for context
            
            # Append the individual review DataFrame to the list
            review_dfs.append(review_df)
        else:
            print(f"No reviews available for POI: {row['name']}")
    else:
        print(f"Failed to retrieve reviews for POI: {row['name']} (Status: {response.status_code})")

# Combine all the individual review DataFrames into one large DataFrame
if review_dfs:
    all_reviews_df = pd.concat(review_dfs, ignore_index=True)

# Display or use the combined DataFrame of reviews
all_reviews_df.head()
```

```python
all_reviews_dfen = all_reviews_df[all_reviews_df['language']=='en']
all_reviews_dfen
```

```python
reviewlist=all_reviews_dfen['text'].tolist()
reviewlist
```

```python
def check_for_synonyms(review, synonyms):
    review_lower = review.lower()  # Convert review to lowercase
    return any(synonym.lower() in review_lower for synonym in synonyms)
goodresult = [check_for_synonyms(review, goodsynonyms_hashed) for review in reviewlist]
sum(goodresult)
```

```python
badresult = [check_for_synonyms(review, badsynonyms_hashed) for review in reviewlist]
sum(badresult)
```
